```python
#filename - wlwwar/XC112689.ogg

file_path = '/content/drive/MyDrive/UCB-MIDS/SEM-2/MACHINE-LEARNING-207/207-Project/data/train/audio_files/thrnig1/XC110335.ogg'
audio, sample_rate = librosa.load(file_path)

# Play the original audio
ipd.Audio(audio, rate=sample_rate)
```

```python
#looks like a standard deviation of 0.01 creates less white noise than a standard deviation of 1 or 0.1
def generate_gaussian_noise(length, mean=0, std=0.003):
    gaussian_noise = np.random.normal(mean, std, length)
    return gaussian_noise

# Generate Gaussian noise
gaussian_noise = generate_gaussian_noise(len(audio))


# Add noise to the original audio
audio_with_noise = audio + gaussian_noise

# Play the audio with noise
print("Audio with Gaussian noise:")
ipd.display(ipd.Audio(audio_with_noise, rate=sample_rate))
```

```python
#Specify the pitch shift factor (in semitones)
pitch_shift_factor = 1 # Adjust the value as needed

#Apply pitch shifting
pitch_shifted_audio = librosa.effects.pitch_shift(audio, sr=sample_rate, n_steps=pitch_shift_factor)

#Play the pitch-shifted audio
ipd.Audio(pitch_shifted_audio, rate=sample_rate)
```

```python
# Flip the audio waveform
flipped_audio = np.flip(audio)

# Play the flipped audio
print("Flipped audio:")
ipd.display(ipd.Audio(flipped_audio, rate=sample_rate))
```
